```python
##processing resolution columns by separating pixel height and width
df['pxl_width'] = df['resolution'].str.split(' x ').str[0]
```

```python
df['pxl_width'] = pd.to_numeric(df['pxl_width'].str.replace('[\D]','',regex=True))
df['pxl_height'] = pd.to_numeric(df['pxl_height'].str.replace('[\D]','',regex=True))
```

```python
df.drop('resolution',axis=1,inplace=True)
```

```python
##processing 'num_cores' column
df['num_cores'] = df['num_cores'].map({'Single Core':1,'Quad Core':2,'Octa Core':3})
```

```python
##processing mobile names
def replace_regex(df,col,chars_to_remove):
    processed = df[col]
    for char in chars_to_remove:
        processed = processed.str.replace(char,'',regex=True)
    return processed
```

```python
df['mobile_name'] = replace_regex(df, 'mobile_name', [' .*'])
```

```python
df['mobile_name'].unique()
```

```python
def convert_words_to_num(df,cat_col):
    return df[cat_col].astype('category').cat.codes
```

```python
df['mobile_name'] = convert_words_to_num(df,'mobile_name')
```

```python
df['mobile_color'] = convert_words_to_num(df,'mobile_color')
```

```python
df['os'] = convert_words_to_num(df,'os')
```

```python
def replace_regex_num(df,col,chars_to_remove):
    processed = df[col]
    for char in chars_to_remove:
        processed = processed.str.replace(char,'',regex=True)
    return pd.to_numeric(processed)
```

```python
df['f_cam'] = replace_regex_num(df,'f_cam',['MP.*'])
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
X = df.drop('mobile_price', axis = 1)
y = df['mobile_price']
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
lr = LogisticRegression(multi_class = 'multinomial', solver = 'sag',  max_iter = 10000)
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
confusion_matrix = metrics.confusion_matrix(y_test,y_pred)
confusion_matrix
```

```python
score = metrics.accuracy_score(y_test, y_pred)
```
